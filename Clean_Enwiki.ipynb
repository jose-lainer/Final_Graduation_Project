## **4. Import keys and vectors**

```
vectors = np.load('EW_w2v_vectors.npy')
print(vectors.shape)


keys = np.load('EW_w2v_keys.npy')
print(keys.shape)

model_keys = keys.tolist()
```

## **5. Clean keys and vectors**

### 5.1 Match keys (439)

```
other_keys = np.load('Visual_att_row_keys_short.npy')
print(other_keys.shape)

m_keys = other_keys.tolist()
```

```
textual_keys = []
textual_vectors = []

for i in range(len(m_keys)): # 0-2182
    for j in range(len(model_keys)): # 0-2514312
        if m_keys[i] == model_keys[j]:
            textual_keys.append(model_keys[j])
            textual_vectors.append(vectors[j])
            break
        else:
            pass
    if i % 100 == 0:
        print(f'{i} of {len(m_keys)}')

clear_output()

print(len(textual_keys))
```

### 5.2 Missing words

```
mw = []

for i in range(len(m_keys)):
    a = 0
    for j in range(len(textual_keys)):
        if textual_keys[j] == m_keys[i]:
            a = 1
            break
    if a == 0:
        mw.append(m_keys[i])

print(f'Missing words: {mw}')
```

## **6. Export keys and vectors**

```
f1 = np.array(textual_keys)
np.save('0_textual_keys_439.npy', f1)

f2 = np.array(textual_vectors, dtype='float32')
np.save('0_textual_features_439.npy', f2)

print('Textual features finished!')
```
