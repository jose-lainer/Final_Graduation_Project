## **3. Prepare dataset**

```
# List with all objects of each image
list_obj = []

# Iterating over each image in the dataset
for i in range(0, len(GenomeData)):
    img_objects = []    # list of objects in each image
    # Iterating over each object in the image and adding to the list
    for obj in range(0, len(GenomeData[i]['objects'])):
        img_objects.append(GenomeData[i]['objects'][obj]['names'][0])
    list_obj.append(img_objects)

# Test
list_obj[10]
```

```
# Find images with no objects
zeros = 0
index = []

# Final list of VisualGenome, objects in each image
vg_list = []

for i in range(0, len(list_obj)):
    if len(list_obj[i]) == 0:
        zeros = zeros + 1
        index.append(i)
    else:
        vg_list.append(list_obj[i])

print('Total of empty images:', zeros, 'of', len(list_obj))
print('Index of empty images:', index)
print('Final list of images', len(vg_list))
```

```
count= 0
for i in range(0, len(vg_list)):
    count = count + len(vg_list[i])

print('Total words:',count)
```

```
VG_SENTENCES = np.array(vg_list)
np.save('VG_sentences.npy', VG_SENTENCES)
```
